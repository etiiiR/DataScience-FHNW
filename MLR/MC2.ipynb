```python
import pandas as pd
kleider = pd.read_csv('kleider.csv', sep=',')
```

```python
# scatterplot aller daten
kleider.plot.scatter(x='groesse', y='gewicht')
```

```python
from sklearn.linear_model import LinearRegression

```

```python
# test ob die variablen signifikant sind
import statsmodels.api as sm
# lineare regression
model = LinearRegression()
model.fit(kleider[['kleidergroesse']], kleider[['gewicht']])
X = kleider[['kleidergroesse']]
y = kleider[['gewicht']]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```
